```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Read Data

```python
netA1_hf = pd.read_csv("./NetA1HF_results.csv")
netA1_ht = pd.read_csv("./NetA1HT_results.csv")
netA1_dt = pd.read_csv("./NetA1DT_results.csv")
```

```python
netA2_hf = pd.read_csv("./NetA2HF_results.csv")
netA2_ht = pd.read_csv("./NetA2HT_results.csv")
netA2_dt = pd.read_csv("./NetA2DT_results.csv")
```

### Utility Function

```python
def plot_single_net(net_df):
    fig1, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

    net_df.plot(x='epoch', y='accuracy', marker='o', linestyle='-', color='b', ax=axes[0] , title='Accuracy over Epochs')
    axes[0].set_xlabel('Epochs')
    axes[0].set_ylabel('Accuracy')
    
    axes[0].grid(True)
    
    net_df.plot(x='epoch', y='loss', marker='o', linestyle='-', color='r', ax=axes[1], title='Loss over Epochs')
    axes[1].set_xlabel('Epochs')
    axes[1].set_ylabel('Loss')
    
    axes[1].grid(True)
    
    plt.show()
    
```

```python
def plot_multi_net(nets_df: list, label_df: list =[]):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    if len(label_df) < len(nets_df):
        label_df += [f"Net_{i}" for i in range(len(label_df)-1, len(nets_df))]
    for i, net_df in enumerate(nets_df):
        ax1.plot(net_df['epoch'], net_df['accuracy'], label=label_df[i], marker='o')
        ax2.plot(net_df['epoch'], net_df['loss'], label=label_df[i], marker='o')
        #ax3.plot(net_df['times'], net_df['accuracy'], label=label_df[i], marker='o')
    ax1.set_title('Accuracy over epochs')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Accuracy')
    ax2.set_title('Loss over epochs')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Loss')
    #ax3.set_title('Accuracy over time')
    #ax3.set_xlabel('Time')
    #ax3.set_ylabel('Loss')
    ax1.grid(True)
    ax2.grid(True)
    #ax3.grid(True)
    ax1.legend()
    ax2.legend()
    #ax3.legend()
```

### Plot Resuts of each network
